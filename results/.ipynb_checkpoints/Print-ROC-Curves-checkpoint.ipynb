```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model_name='IRNV2_noweights_plateau_139_139_e_25'
```

```python
roc_curve_df = pd.read_csv('roc_curve_' + model_name + '.csv')
```

```python
false_positive_rate = roc_curve_df.FalsePositive
true_positive_rate = roc_curve_df.TruePositive
auc = roc_curve_df.AUC[0]
```

```python
plt.figure()
lw = 2
plt.plot(false_positive_rate, true_positive_rate, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
